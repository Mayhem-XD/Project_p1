```python
import numpy as np
import pandas as pd
import folium, requests
from folium.plugins import HeatMap
import requests,json,os
from urllib.parse import quote 
```

```python
df = pd.read_csv('latlng_merge_test.csv')
df.head()
```

```python
df1 = df[['사용월','지하철역','총 승차인원','총 하차인원','lat','lng']].copy()
df1.head()
```

```python
df1 = df1[(df1['사용월'] >= 202301) & (df['사용월'] < 202305)].copy()
df1.head()
```

```python
res = df1.groupby('지하철역')[['총 승차인원','총 하차인원','lat','lng']].agg('mean').reset_index()
res.head()
```

```python
res2 = res[['lat','lng','총 승차인원']]
```

```python

# 지도 생성
m = folium.Map(location=[37.394946, 127.111104], zoom_start=12)
# 데이터
HeatMap(
    data=res2, 
    radius=8,
).add_to(m)
m
#
```
