```python
df1 = pd.read_csv('data/main/1호선.csv')
df2 = pd.read_csv('data/main/경부선.csv')
```

```python
df_1_1 = df1.copy()
df_2_1 = df2.copy()
```

```python
result = pd.concat([df_1_1, df_2_1], axis=0)
result = result.reset_index(drop=True)
result.tail()
```

```python
result.호선명 = '1호선'
```

```python
result.지하철역.nunique(), result.shape, result.지하철역.unique()
```

```python
cols = list(df1.columns)[:3]
target = list(df1.columns)[3:]
cols, target
```

```python
res = result.groupby(cols)[target].agg('sum').reset_index()
res.tail()
```

```python
result.지하철역.nunique(), result.shape, result.지하철역.unique()
```

```python
res.to_csv('data/main/lines/1호선.cvs',index=False)
```
