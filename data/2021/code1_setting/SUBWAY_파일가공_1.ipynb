```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
now = datetime.now()
f_date = now.strftime('%Y%m%d')
df_st = pd.read_csv(f'data/{f_date}_SUBWAY_MONTH.csv', encoding='utf-8', sep=',')
```

- 호선별로 merge 파일 분리 단계
     - 데이터 한번에 정리해야 할 것들 분리 전에 미리 작업하기
    - -> (컬럼정리, 사용일자, 평일/주말 구분 등등..)

```python
############# 요일, 주중/주말 ###################
import datetime
# 일자 기준 주중주말 표시한 컬럼 만들기
week_end = []
for i in range(len(df_st.index)):
    year = int(str(df_st.사용일자.values[i])[:4])
    mm = int(str(df_st.사용일자.values[i])[4:6])
    dd = int(str(df_st.사용일자.values[i])[0:][6:])
    week_end_dict = {0:'주중', 1:'주중', 2:'주중', 3:'주중', 4:'주중', 5:'주말', 6:'주말'}
    week_end.append(week_end_dict[datetime.date(year,mm,dd).weekday()])

# 주중/주말 column 생성
df_st['주중주말'] = week_end
df_st.head(600).tail()
```

```python
######### 파일 저장 - 주중주말 ver ###########
from datetime import datetime
now = datetime.now()
f_date = now.strftime('%Y%m%d')
df_st.to_csv(f'data/{f_date}_SUBWAY_MONTH_주중주말_일자ver.csv',index=False,encoding='utf-8')
```
