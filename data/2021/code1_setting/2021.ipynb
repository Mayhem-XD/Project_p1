```python
import numpy as np
import pandas as pd

df = pd.read_csv('data/CARD_SUBWAY_MONTH_2021.csv',encoding='euc-kr',sep=',')
df.iloc[500:505]
```

- '역명' column -> '지하철역' / 역명 수정

```python
# 경복궁(정부서울청사) -> 경복궁
df.역명 = df['역명'].str.replace('(',' ',regex=True).str.split().str[0]
df.iloc[500:505]
```

```python
# 컬럼 수정
df.rename(columns={'역명': '지하철역','노선명': '호선명'}, inplace=True)
df.drop(columns=['등록일자'], inplace=True)
```

```python
# 요일, 평일주말 출력하는 반복문
import datetime
day_list = []
week_end = []
for i in range(len(df.index)):
    year = int(str(df.사용일자.values[i])[:4])
    mm = int(str(df.사용일자.values[i])[4:6])
    dd = int(str(df.사용일자.values[i])[0:][6:])
    day_dict = {0:'월', 1:'화', 2:'수', 3:'목', 4:'금', 5:'토', 6:'일'}
    week_end_dict = {0:'평일', 1:'평일', 2:'평일', 3:'평일', 4:'평일', 5:'주말', 6:'주말'}
    day_list.append(day_dict[datetime.date(year,mm,dd).weekday()])
    week_end.append(week_end_dict[datetime.date(year,mm,dd).weekday()])
```

```python
# 요일 column 생성
df['요일'] = day_list
df.head(600).tail()
```

```python
# 평일/주말 column 생성
df['평일주말'] = week_end
df.head(600).tail()
```
