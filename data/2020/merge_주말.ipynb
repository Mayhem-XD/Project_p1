```python
# # 사용일자 슬라이싱 -> yyyymm 형태로
ym_list = []
for x in range(len(df.index)):
    ym_list.append(str(df.사용일자.values[x])[:6])
df['사용일자'] = ym_list

# # col수정: 사용일자 --> 사용월
df.rename(columns={'사용일자':'사용월'}, inplace=True)
```
