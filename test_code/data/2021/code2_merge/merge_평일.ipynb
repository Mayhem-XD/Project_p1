````python
df2 = pd.read_csv('data/2021_평일_v2.csv', encoding='utf-8', sep=',')
df2
````

````python
df2['호선명'].nunique(), df2['호선명'].unique()
````

````python
# 흩어진 노선 모아서 -> 대표노선으로 통합(replace로 값 바꿈)
df2.replace({'호선명': ['1호선','경부선','경원선','경인선','장항선']},'1호선' ,inplace=True)
df2.replace({'호선명':'2호선'},'2호선' ,inplace=True)
df2.replace({'호선명':'5호선'},'5호선' ,inplace=True)
df2.replace({'호선명':'7호선'},'7호선' ,inplace=True)
df2.replace({'호선명':'8호선'},'8호선' ,inplace=True)
df2.replace({'호선명': ['3호선','일산선']}, '3호선' ,inplace=True)
df2.replace({'호선명': ['4호선','과천선','안산선']}, '4호선' ,inplace=True)
df2.replace({'호선명': ['9호선','9호선2~3단계']}, '9호선' ,inplace=True)
df2.replace({'호선명': ['수인선','분당선']}, '수인분당선' ,inplace=True)
df2.replace({'호선명': ['경의선','중앙선']}, '경의중앙선' ,inplace=True)

df2['호선명'].nunique(), df2['호선명'].value_counts()
````

````python
del_list = ['공항철도 1호선','우이신설선','경강선','경춘선']
df3 = df2[~(df2['호선명'].str.strip().isin(del_list))]
df3.호선명.value_counts()
````

````python
# 파일 새로 저장
df3 = df3.to_csv('data/2021_평일_merge_0531.csv', index=False, encoding='utf-8')
# 새파일 다시 열기
df = pd.read_csv('data/2021_평일_merge_0531.csv', encoding='utf-8', sep=',')
df
````
