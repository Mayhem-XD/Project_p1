```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import HeatMap
```

```python
df = pd.read_csv('data/main/heatmap_files/lines_4heatmap_주말.csv')
df.tail()
```

```python
target = '승하차총승객수'
start_month = 20210101
end_month = 20211231
line = '수인분당선'
```

```python
df_test = df[(df.호선명 == line)&(df.사용일자 >= start_month)&(df.사용일자 < end_month)].copy()
df_test.drop(columns=['호선명','사용일자'],inplace=True)
df_test = df_test[['지하철역','lat','lng',target]]
df_test.groupby(['지하철역'])[['lat','lng',target]].agg('mean').reset_index()
df_test.drop(columns=['지하철역'],inplace=True)
m = folium.Map(location=[37.394946, 127.111104], zoom_start=11)
data = df_test
HeatMap(data).add_to(m)
title_html = f'<h3 align="center" style="font-size:20px">{line} {start_month} - {end_month} 기간 {target}</h3>'
m.get_root().html.add_child(folium.Element(title_html))
m
```
