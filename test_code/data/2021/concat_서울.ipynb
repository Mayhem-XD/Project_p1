```python
import pandas as pd
```

```python
df_자연 = pd.read_csv('data/서울_자연_final.csv')
df_쇼핑 = pd.read_csv('data/서울_쇼핑_final.csv')
df_문화 = pd.read_csv('data/서울_문화공간_final.csv')
df_명소1 = pd.read_csv('data/서울_문화_명소_final.csv')
df_명소2 = pd.read_csv('data/서울_명소_명소_final.csv')
```

```python
df_자연.head()
```

```python
# 분류 컬럼 추가
df_자연['분류'] = '자연'
df_쇼핑['분류'] = '쇼핑'
df_문화['분류'] ='문화'
df_명소1['분류'] = '명소'
df_명소2['분류'] = '명소'
df_문화.head()
```

```python
df_자연.isna().sum().sum()  # 0
df_쇼핑.isna().sum().sum()    # 0
df_문화.isna().sum().sum()  # 0
df_명소1.isna().sum().sum() # 0
df_명소2.isna().sum().sum() # 0
```

```python
df_명소 = pd.concat([df_명소1, df_명소2])
df_명소.shape
```

```python
df_명소 = df_명소.reset_index(drop=True)
df_명소.to_csv('data/서울_명소1,2_final', index=False)
```
