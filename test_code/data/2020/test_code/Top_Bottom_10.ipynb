```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../main_code_test/data/merged_lines.csv')
df['총 승하차인원'] = df['총 승차인원'] + df['총 하차인원']

# 사용일자 선택
# df1 = df[df['사용월']==202001].copy()
df1 = df[(df['사용월'] >= 202201)&(df['사용월'] <= 202212)].copy()

df1.drop(columns=['호선명','사용월'], inplace=True)
cols = list(df1.columns)[:1]
target = list(df1.columns)[1:]
df1 = df1.groupby(cols)[target].agg('sum').reset_index()

# 표시할 컬럼 선택
df1.sort_values(by='총 승하차인원', inplace=True)

# top10
df1 = df1.head(10)

plt.rcParams['font.family'] = 'Malgun Gothic'

colors = sns.color_palette()
df1.plot.bar(x='지하철역', y='총 승하차인원', color=colors)
plt.title('22년 총 승하차인원 하위10')
plt.savefig('data/graph/2022년도 총 승하차인원.png')
```
