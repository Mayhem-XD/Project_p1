```python
import pandas as pd
```

```python
df1 = pd.read_excel('data/경기도관광쇼핑지현황.xls')
df2 = pd.read_excel('data/경기도문화시설지현황.xls')
df3 = pd.read_excel('data/경기도역사관광지현황.xls')
df4 = pd.read_excel('data/경기도자연관광지현황.xls')
```

```python
df1.head()
```

```python
df1.isna().sum()
```

```python
df1.dropna(inplace=True)
df1.isna().sum()
```

```python
df2.head()
```

```python
df2.dropna(inplace=True)
df2.isna().sum()
```

```python
df3.head()
```

```python
df3.dropna(inplace=True)
df3.isna().sum()
```

```python
df4.head()
```

```python
df4.dropna(inplace=True)
df4.isna().sum()
```

```python
df1['분류'] = '쇼핑'
df1.head()
```

```python
df2['분류'] = '문화'
df2.head()
```

```python
df3['분류'] = '명소'
df3.head()
```

```python
df4['분류'] = '자연'
df4.head()
```

```python
df1.rename(columns={'위치명':'관광지명','정제도로명주소':'도로명주소','정제WGS84위도':'lat','정제WGS84경도':'lng'}, inplace=True)
df1.head()
```

```python
df1 = df1[['관광지명','도로명주소','lat','lng','분류']]
df1.head()
```

```python
df2.rename(columns={'시설명':'관광지명','정제도로명주소':'도로명주소','정제WGS84위도':'lat','정제WGS84경도':'lng'}, inplace=True)
df2.head()
```

```python
df2 = df2[['관광지명','도로명주소','lat','lng','분류']]
df2.head()
```

```python
df3.rename(columns={'관광정보명':'관광지명','정제도로명주소':'도로명주소','정제WGS84위도':'lat','정제WGS84경도':'lng'}, inplace=True)
df3.head()
```

```python
df3 = df3[['관광지명','도로명주소','lat','lng','분류']]
df3.head()
```

```python
df4.rename(columns={'관광정보명':'관광지명','소재주소':'도로명주소','정제WGS84위도':'lat','정제WGS84경도':'lng'}, inplace=True)
df4.head()
```

```python
df4 = df4[['관광지명','도로명주소','lat','lng','분류']]
df4.head()
```

```python
df = pd.concat([df1,df2,df3,df4])
df.head()
```

```python
df.to_csv('data/경기도통합.csv', index=False)
```
